```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv')
df
```

```python
# drop all rows which have only NaN values, because they are useless

df2 = df.dropna(how='all')
df2
```

```python
# drop duplicated entries which have same full_name, email, gender and age

df3 = df2.drop_duplicates(subset={'full_name', 'email', 'gender', 'age'}, keep='first', inplace=False)
df3
```

```python
# remove all customer contacts without email address, because without postal address or phone number 
# there is no possibilty to contact them

df4 = df3.dropna(subset=['email'])
df4
```

```python
# set NaN gender to 'not set' string, it's an default value for missing or non recorded values

df5 = df4
df5.gender = df5.gender.replace(np.nan, 'not set')
df5
```

```python
# reset old string to 0, because every customer has to be over 18 years old
# and set strings of age column to integers for better handling

df6 = df5
df6['age'] = df6['age'].replace('old', '0')
df6['age'] = df6['age'].astype('int')

# remove all rows with age under 18, because customers or newsletter subscribers under 18 years 
# are not relevant

df6 = df6[(df6[['age']] >= 18).all(axis=1)]
df6
```

```python
# set unique id for last entry, according to other entries
# set new index for all entries

df7 = df6
length_of_df = len(df7)
df7.index = range(length_of_df)
df7.iloc[9].id
df7.at[9, 'id'] = 21.0
df7
```
